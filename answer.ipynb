```python
import requests
import json

# Set the endpoint URL
url = "http://localhost:5000/query"

# Prepare the JSON payload
payload = {
    "query": "Toshakhana Case",
    "n_seed": 3
}

# Set headers
headers = {
    "Content-Type": "application/json"
}

# Send the POST request
response = requests.post(url, json=payload, headers=headers)

# Print the complete response text
print(response.text)
```

```python
response.text
```

```python
final_text = ""

# Iterate over each streamed line (each should be a JSON string)
for line in response.iter_lines():
    if line:
        try:
            # Decode the JSON from the line
            token_data = json.loads(line.decode("utf-8"))
            # Extract the "response" field and append it to the final text
            final_text += token_data.get("response", "")
        except json.JSONDecodeError:
            # Handle any lines that aren't valid JSON
            pass

print("Final Human Readable Response:")
print(final_text)
```
